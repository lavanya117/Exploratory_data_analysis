## Exploratory Data Analysis

**Data preparation and cleaning**

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dataset=pd.read_csv('matches.csv')
```

**Total IPL seasons**

```python
dataset['season'].unique()
```

**No of times each team has won the match**

```python
dataset['winner'].value_counts()
```

```python
x=dataset['winner'].value_counts()
x.plot(kind='barh',color='blue')
plt.title("No. of times each team has won the match")
plt.xlabel("Teams")
plt.grid(True)
plt.ylabel("number of wins")
plt.show()
```

**No of matches in each year:**

```python
x=dataset.groupby('season').size()
x
```

```python
plt.pie(x.values,labels=x.index,autopct='%0.2f%%') #decimal values to two place
plt.show()
```

```python
plt.scatter(x.index,x.values,color='red',marker="*",s=200) #x is a dictionary
plt.title("No of matches in each year")
plt.xlabel("year")
plt.ylabel("no of matches")
plt.show()
```

**How many matches were won by a margin of runs**

```python
plt.hist(dataset['win_by_runs'],bins=6,color='orange')
plt.show()
```

**How many matches were won by a margin of wickets**

```python
plt.hist(dataset['win_by_wickets'],bins=6,color="brown")
plt.show()
```

**No of matches happened in each venue**

```python
x=dataset.groupby('venue').size()
x
```

```python
plt.figure(figsize=(10,10)) #figsize(width,heigth) (in inches)
plt.barh(x.index,x.values,color='skyblue')
plt.show()
```

**Boxplot**

```python
plt.boxplot(dataset['win_by_wickets'])
plt.show()
```

**Violinplot**

```python
data=list([dataset['win_by_runs'],dataset['win_by_wickets']])
plt.violinplot(data)
plt.show()
```

**what is the probability of winning a match if the toss was won?**

```python
probability=dataset['toss_winner']==dataset['winner']
t=probability.groupby(probability).size()
t
```

```python
t.index=t.index.map({False:'Lost',True:'Won'})
t
plt.pie(t.values,labels=t.index, autopct="%0.1f%%")
plt.show()
```

**The chances of winning a match if the toss decision was to field and the DL method is applied**

```python
new_dataset=dataset[dataset['dl_applied']==1] #new_dataset contains only the rows where dl_applied is equal to 1
y=new_dataset['toss_decision']=="field"
s=y.groupby(y).size()
s
```

```python
s.index = s.index.map({False: 'Bat',True:'Field'})
```

```python
plt.pie(s.values,labels=s.index,autopct="%0.2f%%", radius=1)
plt.pie([1],radius=0.5,colors=['w'])
plt.show()
```
